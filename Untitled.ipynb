```python
df = pd.read_csv('games.csv')
df['winner'] = df['winner'].replace('white', 0).replace('black', 1)
df = df[df['winner'] == 1]
Mouves = pd.DataFrame(df.moves)

for n in range(len(Mouves)):
    Mouves.moves.iloc[n] = Mouves.moves.iloc[n].split()

Mouves = Mouves.reset_index()
m = Mouves.moves
m1 = [m[i] for i in range(2000)]
m2 = [m[i] for i in range(2000,4000)]
m3 = [m[i] for i in range(4000,6000)]
m4 = [m[i] for i in range(6000,8000)]
m5 = [m[i] for i in range(8000,9107)]
```

```python
def main0(l):   
    partie = run()
    n = 0
    convert = []
    for i in l:
        if n <= len(l):
            deck = deck_chess(partie)
            a,b,c,d = where(i, deck[0],n,partie)
            #print(partie[a][b],i,a,b,c,d), print(n, len(deck))            
            convert.append((a,b,c,d))
        if a == b == c == d == 9:
            return convert
        if a == b == c == d == 0:
            #print('error')
            return False
        partie[c][d]=partie[a][b]
        partie[c][d].x = c
        partie[c][d].y = d
        partie[a][b] = vide
        #affiche(partie)
        n += 1
    return convert
```

```python
convert1 = [main0(m1[i]) for i in range(len(m1)) if main0(m1[i]) != False]
print(True,1)
convert2 = [main0(m2[i]) for i in range(len(m2)) if main0(m2[i]) != False]
print(True, 2)
convert3 = [main0(m3[i]) for i in range(len(m3)) if main0(m3[i]) != False]
print(True, 3)
convert4 = [main0(m4[i]) for i in range(len(m4)) if main0(m4[i]) != False]
print(True, 4)
convert5 = [main0(m5[i]) for i in range(len(m5)) if main0(m5[i]) != False]
print(True,5)
```
